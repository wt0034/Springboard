```python
df['rich_poor'] = df['PersonalIncome'].apply(lambda x: 'Rich' if x > 5 else 'Poor' )
```

```python
df[['PersonalIncome','rich_poor']]
```

```python
drug_last_year= ['marij_year', 'cocaine_year', 'crack_year', 'heroin_year', 'hallucinogen_year','inhalant_year','meth_year'
               ,'painrelieve_year', 'tranq_year', 'stimulant_year', 'sedative_year']
for i in drug_last_year:
    sns.catplot( x= 'rich_poor', hue = i, kind = 'count', data = df)
```

```python
def with_hue(plot, feature, Number_of_categories, hue_categories):
    a = [p.get_height() for p in plot.patches]
    patch = [p for p in plot.patches]
    for i in range(Number_of_categories):
        total = feature.value_counts().values[i]
        for j in range(hue_categories):
            percentage = '{:.1f}%'.format(100 * a[(j*Number_of_categories + i)]/total)
            x = patch[(j*Number_of_categories + i)].get_x() + patch[(j*Number_of_categories + i)].get_width() / 2 - 0.15
            y = patch[(j*Number_of_categories + i)].get_y() + patch[(j*Number_of_categories + i)].get_height() 
            ax.annotate(percentage, (x, y), size = 12)
    plt.show()
    
plt.figure(figsize = (7,5))
for i in drug_last_year:
    ax=sns.countplot('rich_poor', hue = i, data = df)
    with_hue(ax,df.rich_poor,2,2)
```

```python
#H0: two categories are independent

df['rich_poor_cat'] = df['rich_poor'].apply(lambda x: 2 if x=='Rich' else 1)
from scipy.stats import chi2_contingency
p_list = [0.001,0.01,0.05,0.1]
for j in p_list:
    for i in drug_last_year:
        Independent_list = []
        Not_Independent_list = []
        stat, p, dof, expected = chi2_contingency([df[i], df['rich_poor_cat']])
        if p>j:
            Independent_list.append(i)
        else:
            Not_Independent_list.append(i)
    print('When p threshold is at ', j, '-- Independent :', Independent_list)
    print('When p threshold is at ', j ,'-- Not Independent ;', Not_Independent_list)
```
